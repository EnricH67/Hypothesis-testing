Agafa un conjunt de dades de tema esportiu que t'agradi i selecciona un atribut del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.stats import ttest_ind
```

+ *Sobre el data set amb el que vinc treballant ("tenis_df"), vull fer un contrast d'hipòtesis, en el que la meva hipòtesi nula
  serà que la durada dels partits de la ronda ("round") de "Quarterfinals", té la mateixa distribució que la durada dels 
  partits del data set original, on estan recollides totes les rondes (des de "1st Round" fins a "The Final)*
  
  *Per tant, la meva hipòtesi alternativa serà que la durada dels partits del round "Quarterfinals" té una distribució diferent 
  a la del total de rondes*

```python
tenis_df = pd.read_csv('Match.csv')
tenis_df
```

+ *Comprovo si hi ha null values en el data set*

```python
tenis_df['match_minutes'].isnull().sum()
```

+ *I em desfaig de les observacions que contenen aquests null values*

```python
tenis_df = tenis_df.dropna()
tenis_df
```

+ *Comprovo que ja no n'hi ha cap de null value

```python
tenis_df['match_minutes'].isnull().sum()
```

```python
tenis_df['match_minutes'].mean()
```

```python
tenis_df['match_minutes'].describe()
```

```python
tenis_quarter_df = tenis_df[(tenis_df['round'] == 'Quarterfinals')]
tenis_quarter_df
```

```python
tenis_quarter_df['match_minutes'].isnull().sum()
```

```python
tenis_quarter_df['match_minutes'].mean()
```

```python
tenis_quarter_df['match_minutes'].describe()
```

```python
stat, p = stats.ttest_ind(tenis_quarter_df['match_minutes'], tenis_df['match_minutes'])
print('stat =%.3f, p =%.5f' % (stat, p))
if p > 0.05:
    print ('No podem rebutjar hipòtesi nula')
else:
    print('Rebutgem hipòtesi nula')
```

+ *El p-value obtingut és inferior al valor d'alfa proposat (0.05) i per tant es rebutja l'hipòtesi nula.És a dir la 
   durada dels partits de "Quarterfinals" tenen una distribució diferent a la de la durada  de la totalitat rondes*

+ *__Un cop realitzat l'exercici em dono compte de que les observacions dels partits de "Quarterfinals", són presents tant en el 
   data set original (on hi són totes les rondes) com en el en el sub data set on he filtrat els partits de "Quarterfinals"__*
   
  *__És a dir, les tinc tant en les dades amb que formulo la hipòtesi nula, com a les dades amb que formulo la hipòtesi 
   alternativa__*

+ *I genero un sub data set amb un sample aleatori de 100 observacions de l'original, on només hi són els partits del French

   *Ho = els partits dels tornejos que no son del French Open duren el mateix que els del French Open*

Continua amb el conjunt de dades de tema esportiu que t'agradi i selecciona tres atributs del conjunt de dades. Calcula el p-valor i digues si rebutja la hipòtesi nul·la agafant un alfa de 5%.

+ *Per aquest estudi he pensat en estudiar el comportament de l'atribut "match_minutes" de 3 tornejos diferents
   en segona ronda ("2nd Round")*

  *Per exemple Wimbledon, "Australia" i "US Open".Per tant  generaré 1 sub data set amb una mostra aleatòria de 30 observacions
   per cada un dels tornejos seleccionats*
   
   *La meva Hipòtesi nula és:* 
   
   *H0 = La durada dels partits ("match_minutes") de segona ronda ("round" == "2nd Round") dels 3 tornejos mencionats és igual
        
   *La meva Hipòtesi alternativa és:*
   
   *H1 = La durada dels partits mencionats del 3 tornejos és diferent.* 
        
   

```python
tenis_wimbledon_2ndround_df= tenis_df[(tenis_df['tournament']== "Wimbledon") & (tenis_df['round'] == "2nd Round")].sample(30)
tenis_wimbledon_2ndround_df
```

```python
tenis_australia_2ndround_df=tenis_df[(tenis_df['tournament']== 'Australian Open') & (tenis_df['round']=="2nd Round")].sample(30)
tenis_australia_2ndround_df
```

```python
tenis_usopen_2ndround_df = tenis_df[(tenis_df['tournament'] == "US Open") & (tenis_df['round'] == "2nd Round")].sample(30)
tenis_usopen_2ndround_df
```

```python
from scipy.stats import f_oneway
```

```python
stat, p = f_oneway(tenis_wimbledon_2ndround_df['match_minutes'], tenis_australia_2ndround_df['match_minutes'], tenis_usopen_2ndround_df['match_minutes'])
```

```python
print('stat =%.3f, p =%.15f' % (stat, p))
if p > 0.05:
    print ('No podem rebutjar hipòtesi nula')
else:
    print('Rebutgem hipòtesi nula')
```

+ *__CONCLUSIÓ: per un nivell de confiança alpha = 0.005 amb les 3  mostres aleatòries de 30 observacions, ens dona un p-value   > alpha; per lo qualNO podem rebutjar la hipòtesi nula.__*

+ *__És a dir, no podem rebutjar que els partits de segona ronda dels 3 tornejos evaluats, tenen una durada igual__*

+ *__Veiem qué passaria si agafem una mostra aleatòria més gran__*

```python
tenis_wimbledon_2ndround_df2= tenis_df[(tenis_df['tournament']== "Wimbledon") & (tenis_df['round'] == "2nd Round")].sample(120)
tenis_wimbledon_2ndround_df2
```

```python
tenis_australia_2ndround_df2=tenis_df[(tenis_df['tournament']=='Australian Open') & (tenis_df['round']=="2nd Round")].sample(120)
tenis_australia_2ndround_df2
```

```python
tenis_usopen_2ndround_df2 = tenis_df[(tenis_df['tournament'] == "US Open") & (tenis_df['round'] == "2nd Round")].sample(120)
tenis_usopen_2ndround_df2
```

```python
stat, p = f_oneway(tenis_wimbledon_2ndround_df2['match_minutes'], tenis_australia_2ndround_df2['match_minutes'], tenis_usopen_2ndround_df2['match_minutes'])
```

```python
print('stat =%.3f, p =%.15f' % (stat, p))
if p > 0.05:
    print ('No podem rebutjar hipòtesi nula')
else:
    print('Rebutgem hipòtesi nula')
```

+ *__CONCLUSIÓ: per un nivell de confiança alpha = 0.005 amb les 3  mostres aleatòries de 120 observacions, ens dona un p-value   < alpha; per lo qual rebutgem la hipòtesi nula.__*

+ *__És a dir, acceptem que els partits de segona ronda dels 3 tornejos evaluats, tenen una durada diferent__*
